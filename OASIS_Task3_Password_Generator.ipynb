```python
import string
import random
print("\t\t\t\t\t\tPASSWORD GENERATOR\n")
print("\n\n")
choice = True
while(choice):
    length = int(input("Enter the length of the password : "))
    if length == 0 or length<0:
        print("Password cannot be generator. Length cannot be 0 or negative.")
    else:
        letters = string.ascii_letters
        digits = string.digits
        punct = string.punctuation
        n1 = int(input("Specify your character type \nPress 1 for letters.\nPress 2 for numbers.\nPress 3 for symbols.\nPress 4 for all."))
        if n1 == 1:
            string_length = len(letters)
            password = []
            for i in range(length):
                n = random.randint(0, string_length - 1)
                password.append(letters[n])

            password = "".join(password)
            print("The Password generated is : ", password)
        elif n1 == 2:
            string_length = len(digits)
            password = []
            for i in range(length):
                n = random.randint(0, string_length - 1)
                password.append(digits[n])

            password = "".join(password)
            print("The Password generated is : ", password)
        elif n1 == 3:
            string_length = len(punct)
            password = []
            for i in range(length):
                n = random.randint(0, string_length - 1)
                password.append(punct[n])

            password = "".join(password)
            print("The Password generated is : ", password)
        elif n1 == 4:
            s = letters+digits+punct
            string_length = len(s)
            password = []
            for i in range(length):
                n = random.randint(0, string_length - 1)
                password.append(s[n])

            password = "".join(password)
            print("The Password generated is : ", password)
        else:
            print("Wrong character type entered.")
    ch = input("Do you want to generate another password ? (Y/y) or (N/n) ")
    if ch=="Y" or ch=="y":
        choice = True
    else:
        choice = False
        
print("\nThank you!!")
```
